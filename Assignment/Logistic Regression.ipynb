## Step3: Separate Data into X and Y feature
    Y = 'price_range'
    X = Remaining all the features

```python
X = df.drop(columns='price_range')
Y = df[['price_range']]
```

```python
X.head()
```

```python
Y.head()
```

## Step4: Data Preprocessing and Data Cleaning

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
con =list(X.columns[X.dtypes != 'object'])
```

```python
print(con)
```

```python
pre = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler()
).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

## Step5: Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_pre,Y,train_size=0.75,random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

## Step6: Build the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.classes_
```

```python
ypred_train = model.predict(xtrain)
```

```python
yprob_train = model.predict_proba(xtrain)
```

```python
yprob_train[0]
```

```python
ypred_train[:5]
```

## Model Evaluation

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

## Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model,xtest,ytest)
```

## Classification Report

```python
from sklearn.metrics import classification_report
ypred_test =model.predict(xtest)
ypred_test[:5]
```

```python
print(classification_report(ytest,ypred_test))
```
